```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
#LAB01
sin_sig = thinkdsp.SinSignal(freq=7003, amp=3, offset=0)
sin_sig.plot()
thinkplot.config(xlabel = 'Time (s)')
```

```python
#LAB02
#trianglesignal(freq)
signal = thinkdsp.TriangleSignal(3)
duration = signal.period*3
segment = signal.make_wave(duration, framerate=10000)
segment.plot()

kawwave=signal.make_wave(duration,framerate=10000)
kawwave.apodize()
kawwave.make_audio

spectrum = kawwave.make_spectrum()
spectrum.plot()
```

```python
signal = thinkdsp.SquareSignal(3)
duration = signal.period*3
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
sqwave=signal.make_wave(duration,framerate=10000)
sqwave.apodize()
sqwave.make_audio()

```
